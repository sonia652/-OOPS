#QUESTION 8 .How is encapsulation achieved in Python?

#.ncapsulation in Python is achieved by restricting access to variables and methods to protect data from accidental modification. It involves binding data and methods together and controlling their access.

Here‚Äôs how encapsulation is achieved in Python:

Using access specifiers:

Public members (default): accessible from anywhere.

Protected members (single underscore _var) : intended to be accessed only within the class and its subclasses (still accessible outside, but by convention treated as protected).

Private members (double underscore __var): not accessible directly outside the class due to name mangling.

```python
#QUESTION 2.  Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both.

#.from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Circle class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

# Rectangle class
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Testing
c = Circle(5)
r = Rectangle(4, 6)

print("Circle area:", c.area())       # Output: Circle area: 78.54...
print("Rectangle area:", r.area())    # Output: Rectangle area: 24
```

```python
#QUESTION 3. . Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute.

#. # Parent class
class Vehicle:
    def __init__(self, type):
        self.type = type

# Child class of Vehicle
class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

# Child class of Car (multi-level inheritance)
class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery

# Testing
ecar = ElectricCar("Four Wheeler", "Tesla", "100 kWh")
print("Type:", ecar.type)          # Output: Type: Four Wheeler
print("Brand:", ecar.brand)        # Output: Brand: Tesla
print("Battery:", ecar.battery)    # Output: Battery: 100 kWh
```

```python
#QUESTION 4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
Sparrow and Penguin that override the fly() method.


#. # Base class
class Bird:
    def fly(self):
        print("Birds can fly")

# Derived class Sparrow
class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky")

# Derived class Penguin
class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly")

# Testing polymorphism
birds = [Sparrow(), Penguin()]

for bird in birds:
    bird.fly()
```

```python
#QUESTION 5.Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance.

#.
class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance   # private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient balance or invalid amount")

    def check_balance(self):
        print(f"Current balance: {self.__balance}")

```

```python
#QUESTION 6.  Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
and Piano that implement their own version of play().

#.
# Base class
class Instrument:
    def play(self):
        print("Instrument is playing")

# Derived class Guitar
class Guitar(Instrument):
    def play(self):
        print("Guitar is playing melodious tunes")

# Derived class Piano
class Piano(Instrument):
    def play(self):
        print("Piano is playing classical music")

# Testing runtime polymorphism
instruments = [Guitar(), Piano()]

for instrument in instruments:
    instrument.play()
```

```python
#QUESTION 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers.

#.
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Testing
print("Addition:", MathOperations.add_numbers(10, 5))       # Output: Addition: 15
print("Subtraction:", MathOperations.subtract_numbers(10, 5))  # Output: Subtraction: 5
```

```python
#QUESTION 10 .Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors.


#.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Testing
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2  # uses __add__

print("v1 + v2 =", v3)  # Output: v1 + v2 = (6, 8)
```

```python
#QUESTION 12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
the average of the grades.

#.
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # list of grades

    def average_grade(self):
        if self.grades:
            return sum(self.grades) / len(self.grades)
        else:
            return 0

# Testing
s = Student("John", [85, 90, 78, 92])
print(f"{s.name}'s average grade:", s.average_grade())  # Output: John's average grade: 86.25

```

```python
#QUESTION 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method

#.
from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Derived class Cow
class Cow(Animal):
    def sound(self):
        print("Cow says Moo")

# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        print("Sheep says Baa")

# Testing
c = Cow()
s = Sheep()

c.sound()   # Output: Cow says Moo
s.sound()   # Output: Sheep says Baa
```

```python
#QUESTION 17 . . Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
returns a formatted string with the book's details.


#.
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

# Testing
b = Book("The Alchemist", "Paulo Coelho", 1988)
print(b.get_book_info())  # Output: 'The Alchemist' by Paulo Coelho, published in 1988
```

```python
#QUESTION 18.Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms

#.
# Base class
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

# Derived class
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

# Testing
m = Mansion("123 Elite Street", 50000000, 10)
print("Address:", m.address)               # Output: 123 Elite Street
print("Price:", m.price)                   # Output: 50000000
print("Number of rooms:", m.number_of_rooms)  # Output: 10

```
